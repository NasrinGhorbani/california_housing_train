**Score of Regression Model without one feature:**

```python
x_x = df.drop(df.columns[8], axis=1)

model_i = LinearRegression()
score = []
n = 1

for i in range(0,8):
  x_i = x_x.drop(df.columns[i], axis=1)
  model_i.fit(x_i ,y)
  score_i = model_i.score(x_i ,y)
  score.append(score_i)
  max_score = max(score)
  max_index = score.index(max_score)
  max_index = int(max_index)
  feature_loss = df.columns[i]
  print(n)
  n = n+1
  print('Score of Regression Model whitout','"', feature_loss,'"', 'is:', score_i)

print('')
print('**The maximum score of Regression Models is:', max_score, 'which is blongs to Model without','"', df.columns[max_index],'"','feature.')
```

**Score of Regression Model without two features:**

```python
n = 1
score = []
max_score = float('-inf')
max_i = None
max_j = None

for j in range(0,8):
  x_j = x_x.drop(df.columns[j], axis=1)
  for i in range(0,8):
    if i==j:
      break

    x_i = x_j.drop(df.columns[i], axis=1)
    model_i.fit(x_i ,y)
    score_i = model_i.score(x_i ,y)
    if score_i > max_score:
      max_score = score_i
      max_i = i
      max_j = j
    
    feature_loss_i = df.columns[i]
    feature_loss_j = df.columns[j]
    print(n)
    n = n+1
    print('Score of Regression Model whitout','"', feature_loss_j,'"','and','"', feature_loss_i,'"', 'is:', score_i)
  
print('')
print('**The maximum score of these Regression Models is:', max_score, 'which is blongs to Model without','"', df.columns[max_j],'"','and','"', df.columns[max_i],'"','features.')
```
